# Animate anything

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from lipstick import GifMaker

with GifMaker("sample.gif") as g:
    
    for i in range(30):
        fig, ax = plt.subplots(figsize=(2, 2), dpi=150)
        ax.imshow(np.random.rand(3, 3))
        ax.text(2, 2, i, ha='center', va='center')
        ax.set(xticks=[], yticks=[])
        
        g.add(fig)
        
g.show()
```

## Updating figure witihin a loop

```python
from lipstick import update_fig

fig, ax = plt.subplots(figsize=(2, 2), dpi=150)
for i in range(10):
    ax.imshow(np.random.rand(3, 3))
    ax.text(2, 2, i, ha='center', va='center')
    ax.set(xticks=[], yticks=[])
    
    update_fig(fig, ax, sleep=.5)
```

## Updating figure with multiple axes

```python
from lipstick import update_fig

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(4, 2), dpi=150)
for i in range(10):
    
    ax1.imshow(np.random.rand(3, 3))
    ax1.text(2, 2, i, ha='center', va='center')
    ax1.set(xticks=[], yticks=[])
    
    ax2.imshow(np.random.rand(3, 3))
    ax2.text(2, 2, i, ha='center', va='center')
    ax2.set(xticks=[], yticks=[])
    
    update_fig(fig, (ax1, ax2), sleep=.5)
```
